```python
from autoscraper import AutoScraper
import pandas as pd
```

```python
# Initialize the AutoScraper
scraper = AutoScraper()
```

```python
url = "https://www.flipkart.com/flipkart-smartbuy-1000-w-dry-iron/product-reviews/itmf3kcr58zcbh6p?pid=IRNF28ETGB7HHRWE&lid=LSTIRNF28ETGB7HHRWEYGG9WM&marketplace=FLIPKART"
wanted_list = ["Worth every penny","Thanks Flipkart, Good Product Good Quality , I am Happy With this Product & Satisfied","Shekhar Kumar","5"]


scraper.build(url, wanted_list)


scraper.save('flipkart-reviews')


all_results = []

for page in range(1, 110):
    page_url = f"{url}&page={page}"
    page_results = scraper.get_result_similar(page_url, grouped=True)
    all_results.append(page_results)


all_results

flattened_results = {
    'UserName': [],
    'Rating': [],
    'Review': []
}
index_n = []
for page_data in all_results:
    for index, (key, value) in enumerate(page_data.items()): 
        index_n.append(key)
        if index == 0:
            flattened_results['Review'].append(value)
        if index == 1:
            flattened_results['UserName'].append(value)
        if index >= 2:
            flattened_results['Rating'].append(value)
            
flattened_results
```

```python

df = pd.DataFrame({
    'Username': [item for sublist in flattened_results['UserName'] for item in sublist],
    'Review': [item for sublist in flattened_results['Review'] for item in sublist],
    'Rating': [item[0] for item in flattened_results['Rating']]
})
```
